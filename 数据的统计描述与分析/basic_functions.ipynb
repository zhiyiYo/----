## 导入包和模块

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats.distributions import _continuous_distns as cd
```

## 基本统计

```python
data = pd.read_excel('resource\\身高体重表.xlsx')
res = stats.describe(data)
```

## 求中心矩

```python
# 求n阶中心矩
two_order_central_moment = stats.moment(data, 2)
# 求偏度(3阶中心矩)
stats.kurtosis(data)
# 求峰度(4阶中心矩)
stats.skew(data)
```

##  产生随机数据

```python
norm_data = cd.norm.rvs(loc=5, scale=3, size=100)   # 正态分布
chi2_data = cd.chi2.rvs(df=3, size=100)             # 自由度n=3的 χ^2分布
expond_data = cd.expon.rvs(scale=1/5)               # λ=5的指数分布
```

## 计算概率密度值

```python
chi2_x = np.linspace(0, 10, 200)
chi2_y = cd.chi2.pdf(chi2_x, df=3)
plt.plot(chi2_x, chi2_y)
```

## 计算累计概率分布值

```python
norm_x = np.linspace(-10, 10, 200)
norm_y = cd.norm.cdf(norm_x, loc=0, scale=1)
plt.plot(norm_x, norm_y)
```

## 计算分布的均值和方差

```python
print('χ^2(5)的均值：', cd.chi2.mean(df=5), '方差：', cd.chi2.var(df=5))
```
